# Algorithms vs Programs
Although both are defined as "step by step procedures to solve a computational program," the difference between them is not the *what*, but the *when* we write them, and *how*.

**Algorithms** are written in the design phase of the Software Development cycle. They're usually written on paper, although anything could be used. It depends on people with Domain Knowledge, and can be written on any language available, although usually they're written in English or using mathematical notation. They're independent of hardware and software; it doesn't matter what operating system you use. Algorithms are written to be analysed.

\* Note: Although any language can be used, we tend to use C language or something that most people will understand and know.

**Programs** are written during the impelemntation phase, and are writter by programmers, using programming language. They are dependent and reliant on hardware, software, and the operating system (it's not the same writing code for Linux that for Windows). Programs are written to be tested.

## Analysis vs Testing
**Priori Analysis**: Analyzing an algorithm, with the aims of finding the time and space function.


**Posteriori Testing**: Testing is performed on programs, and it watches times and bytes.

## Characteristics of Algorithms

**Input**: An algorithm must take 0 or more input.

**Output**: An algorithm must generate at least one output, otherwise it's useless: it should generate someething.

**Definiteness:** Every statement/step in the algorithm should be independent and should be clear. For example, $\sqrt{-1}$ it's an unknown value, and it's not clear.

**Finiteness:** Since an algorithm is just a function, an algorithm should be limited, and stop and terminate at some point, no matter how many steps/statements it has.

**Effectiveness:** An algorithm should not have unnecessary statements, so every step should have meaning and be essential for the algorithm.

## How to write an Algorithm
Consider the following algorithm, meant to swap two items:

**Algorithm Description:**

**Pseudocode:**
```
Algorithm Swap(a, b)
Begin
    temp = a
    a = b
    b = temp
End
```

**Correctness:** The algorithm is correct because it assigns
`a` to a temporal variable before changing its value to be the same as `b`, and then assigns b to be the temporal variable, ensuring that the values of `a` and `b` are preserved.

**Time analysis:**
```
Algorithm Swap(a, b)
Begin                    
    temp = a             -> 1
    a = b               -> 1
    b = temp            -> 1
End
```

Every step in the algorithm takes constant/O(1)/1 time, and since it has 3 steps, the function of time is determined by $T(n) = 3$.

**Space analysis:**
It stores 3 variables, `a`, `b`, `c`, , with all of them storing only one value at the time so the total space is given be the function $S(n) = 3$.

### How to Analyze an Algorithm

1. **Time**: The first thing being analyzed is time. Any procedure must be time efficient. Time is always written in the form of function.

2. **Space**: Since an algorithm will turn into a program that will run in a machine, we must know how much storage/space it takes in the machine.

3. **Data Transfer/Network Consumption**: How much data is going to be transferred.

4. **Power Consumption**: How much power of the machine is being consumed when running the program.

5. **CPU Registers**: How many CPU registers is the algorithm/program consuming.

## The Frequency Count Method
Consider the following algorithms:

> **Algorithm Description:**
>
> **Pseudocode:**
> ```
> Algorithm Sum(A, n)
> BEGIN
>    s = 0
>    for i = 0 while i < n
>        s = s + A[i]
>    return s
> END
> ```
>
> **Correctness:**
>
> **Time Analysis:**
> > The time taken by an algorithm can be known by assigning a unit of time to each statement. If a statement is repeating for a number of times, the frequency of execution of the satement will be calculated using the **Frequency Count Method**.
> ```
> Algorithm Sum(A, n)      
> BEGIN
>    s = 0                      -> 1
>    for i = 0 until i = n      -> n+1
>        s = s + A[i]           -> n
>    return s                   -> 1
> END
> ```
>
> The function of time is given by $T(n) = 2n + 3$.
> 
> **Space Analysis:**
> Uses the variables `A` (size n), `n` (size 1), `s` (size 1), and `i` (size 1), so the function of space is given by $S(n) = n + 3$.


> **Algorithm Description:**
> 
> **Pseudocode:**
> ```
> Algorithm Add(A, B, n)
> BEGIN
>    for i = 0 while i < n
>        for j = 0 until j = n
>            C[i, j] = A[i, j] + B[i, j]
> END
> ```
>
> **Correctness:**
>
> **Time Analysis:**
> ```
> Algorithm Add(A, B, n)
> BEGIN
>    for i = 0 while i < n                -> n + 1
>        for j = 0 until j = n            -> n(n + 1)
>            C[i, j] = A[i, j] + B[i, j]  -> n(n)
>END
>```
>
> The function of time is given by $T(n) = 2n^2 + 2n + 1$.
> 
> **Space Analysis:**
Uses the variables `A` (size n$^2$), `B` (size n$^2$), `C` (size n$^2$), `n` (size 1), `i` (size 1), and `j` (size 1), so the function of time is given by $T(n) = 3n^2 + 3$.

> **Algorithm Description:**
>
> **Pseudocode:**
> ```
> Algorithm Multiply(A, B, n)
> BEGIN
>    for i = 0 while i < n
>        for j = 0 while j < n
>            C[i,j] = 0
>            for k = 0 while k < n
>                C[i,j] - C[i,j] + A[i,k] * B[k,j]
> END
> ```
>
> **Time Analysis:**
> ```
> Algorithm Multiply(A, B, n)
> BEGIN
>    for i = 0 while i < n                           -> n + 1
>        for j = 0 while j < n                       -> n(n + 1)
>            C[i,j] = 0                              -> n(n)(1)
>            for k = 0 while k < n                   -> n(n)(n + 1)
>                C[i,j] - C[i,j] + A[i,k] * B[k,j]   -> n(n)(n)
> END
> ```
> The function of time is given by $T(n) = 2n^3 + 3n^2 + 2n + 1$.
>
> **Space Analysis**: We use variables `A`, `B`, and `C` (size n^2), and `n`, `i`, `j`, `k` (size 1). The function of space is given by $S(n) = 3n^2 + 4$.
