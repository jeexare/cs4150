# Algorithm Analysis

When designing, writing, testing and developing algoriothms, the number one question asked is: "how fast is it?" This question is not easy to answer, because "fast" is tricky to answer. After all, how fast something is depends on:

- Input size (it isn't the same thing sorting 10 items vs sorting 1 million items)
- Input pattern (what if the list is already sorted vs reverse sorted)
- Machine (supercomputer vs old laptop)

## The Solution: A Standard Framework

To be able to find out how "fast" something is and explain it to others, we can talk in terms of a function $T(n)$, where:

- $n$ = input size (like number of elements in an array)
- $T(n)$ = worst-case running time for inputs of size $n$

With this, we ignore the best case since it's usually meaningless, as well as specific machine details, and instead focus only on the worst case.
