# Algorithm Design Framework

When asked to design an algorithm, design first a **brute-force** algorithm, and ask the following questions:

1. Can we make the problem smaller by splitting the input in half?
2. After splitting do I need BOTH halves or can I thrown one away?
3. What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?
4. What do we need to return for the recursion to work at the next level?
5. Is the input sorted or do we have indices to work with?
6. How many recursive calls do we make? On what?
7. After the recursive calls return, what do I need to get the final answer?
8. What's my recurrence, and can I solve it with Master Theorem?

To help answer these questions, draw a diagram.

## Practice problems
### FindUniqueIndex
**Problem:** Given a sorted array A of n distinct integers, some of which may be negative, find an index i such that A[i] = i. If no such index exists, return -1.

**Example:** A = [-3, 0, 2, 4, 7, 9] The answer is i = 4 because A[4] = 4.

First, we can design a **brute-force** approach:

We loop through the array once, checking if A[i] is negative, then we go to the next one because i obviously not negative, and if we find a A[i] = i, we return i, otherwise we return -1.

This would O(n).

**Can we make the problem smaller by splitting the input in half?**
Yes, if we have an array |A| / 2, we can recurse on it and it can still work.

**After splitting do I need BOTH halves or can I thrown one away?**
I can throw one away, because the array is sorted, so if mid < A[mid], we know the answer is either on the left, or there's not answer, and viceversa.

**What's the smallest possible input where the answer is obvious or can be brute-forced with minimal cost?**
When there's only one element (i == j), we can check if A[i] == i.

**What do we need to return for the recursion to work at the next level?**
The index i, or -1.

**Is the input sorted or do we have indices to work with?**
Sorted.

**How many recursive calls do we make? On what?**
We can make 1 on only half the array each time, until the base case.

**After the recursive calls return, what do I need to get the final answer?**
We don't need to combine anything, just return the recursive call, since once it reaches the base case, we have our answer.

**What's my recurrence, and can I solve it with Master Theorem?**
$T(n) = T(n/2) + O(1)$, yes it can be solved with Master Theorem.

```
Algorithm FindUniqueIndex(A, i, j)
Input: A sorted array A with n distinct integers, a starting index i, an ending index j
Output: A index x such that A[x] == x; or -1

if i == j:
    if A[i] == i:
        return i
    else:
        return -1

mid = (i + j)/2

if A[mid] == mid:
    return mid
else if A[mid] < mid:
    return FindUniqueIndex(A, i, mid - 1)
else:
    return FindUniqueIndex(A, mid + 1, j)
```

### BinarySearch

**Problem:** Given a sorted array of n distinct positive integers, find if a number x exists in the array. Return true or false.

**Example:** A = [1, 3, 5, 7, 9, 11], x = 7 Output: true

**Brute-force** would have us go through the whole array once, comparing each index to x, and returning true if found, false if not. This is O(n).

**Can we make the problem smaller by splitting the input in half?**
Yes, we can divide the array and half and it would still preserve it.

**After splitting do I need BOTH halves or can I thrown one away?**
Since the array is sorted, we only need one half. We can choose the middle index as the pivot, compare with the target. We either find it right away, or our pivot is bigger than the target (so we recurse on the left), or is smaller (so we recurse on the right.

**What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?**
If there's only one element in the array, it's either the answer or it's not.

**What do we need to return for the recursion to work at the next level?**
We have to return true or false.

**Is the input sorted or do we have indices to work with?**
Sorted.

**How many recursive calls do we make? On what?**
One, on half the array.

**After the recursive calls return, what do I need to get the final answer?**
We just need to return the answer of the recursive call and that's the final answer.

**What's my recurrence, and can I solve it with Master Theorem?**
It's $T(n) = T(n/2) + O(1)$. Yes, master theorem can be used.

``` 
Algorithm BinarySearch(A, i, j, x)
Input: A sorted array of n positive integers, a starting index i, an ending index j, a target number x
Output: True or false depending if we find the target in the array

if i == j:
    if A[i] == x:
        return True
    else:
        return False

mid = (i+j)/2

if A[mid] == x:
    return True
else if A[mid] > x:
    return BinarySearch(a, i, mid - 1, x)
else:
    return BinarySearch(a, mid + 1, j, x)
```

### CountNegatives

**Problem:** Given a sorted array A of n integers, count how many elements are negative. Design an O(log n) algorithm.

**Example:** A = [-5, -3, -1, 2, 4, 7, 9] Output: 3

A **brute-force** would have us go through the array, check if the element is negative, and then count. When we reach 0 or a positive element, we return out count. This is O(n)

**Can we make the problem smaller by splitting the input in half?**
Yes, we can divide the array and still preserve it.

**After splitting do I need BOTH halves or can I thrown one away?**
Since it's sorted, we can throw one half away. We can make the middle index a pivot, and check it's neighbors. If A[mid] < 0 AND A[mid + 1] >= 0, then we return mid. If A[mid] > 0, we recurse on the left side. Else, we recurse on the right side.

**What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?**
Only one element: if is negative, we return its index, otherwise its index - 1

**What do we need to return for the recursion to work at the next level?**
See previous question.

**Is the input sorted or do we have indices to work with?**
Sorted.

**How many recursive calls do we make? On what?**
One, only on half the array.

**After the recursive calls return, what do I need to get the final answer?**
The final recursive call will return the correct answer. No need for combine step.

**What's my recurrence, and can I solve it with Master Theorem?**
It should be $T(n) = T(n/2) + O(1)$, and it can be solved with Master Theorem.

```
Algorithm CountNegatives(A, i, j)
Input: A sorted array of integers A, a starting index i, an ending index j
Output: The total count of negative integers in A

if A[i] >= 0:
    return 0
if A[j] < 0:
    return j

if i == j:
    if A[i] < 0:
        return i
    else:
        return i - 1

mid = (i + j)/2

if A[mid] < 0 AND A[mid + 1] >= 0:
    return mid
else if A[mid] >= 0:
    return CountNegatives(A, i, mid - 1)
else:
    return CountNegatives(A, mid + 1, j)
```

### 4.

**Problem:** Given two sorted arrays A and B each of size n, find the median of all 2n elements combined in O(log n) time.

**Example:** A = [1, 3, 5], B = [2, 4, 6] Combined sorted: [1, 2, 3, 4, 5, 6] Output: 3 (or 4, middle elements)

A **brute-force** algorithm would have us combined both arrays into a sorted array, and then return the middle element. This approach would be $O(n log n)$.

**Can we make the problem smaller by splitting the input in half?**
Since we have two arrays, we could divide both in half each time.

**After splitting do I need BOTH halves or can I thrown one away?**
I can throw one half of each array each time. Just find the median of each array, if Median(A) < Median(B), throw away A1 and B2. If Median(A) > Median(B) throw away A2 and B1. If Median(A) = Median(B), just return one of the medians.

**What's the smallest possible input where the answer is obvious or can be brute-forced with minimal cost?**
When the size of each array is just 1, just return the lower median.

**What do we need to return for the recursion to work at the next level?**
Return the lower median.

**Is the input sorted or do we have indices to work with?**
Sorted.

**How many recursive calls do we make? On what?**
We make one recursive calls, on one half of each array.

**After the recursive calls return, what do I need to get the final answer?**
Return the lower median of the two medians.

**What's my recurrence, and can I solve it with Master Theorem?**
$T(n) = T(n/2) + O(1)$, and the Master Theorem can be used.

### 5.

**Problem:** Given an array of n integers, find the closest number to zero. Design an O(n) algorithm using D&C.

**Example:** A = [4, -1, 8, -6, 3, -2] Output: -1

A **brute-force** algorithm would have us go through each single element and store it as they get closer to zero. I.e., looping through A, output = A[0] = 4 at initialization, with distance = 0 + |4| = -4. At A[1] = -1, distance1 = 0 + |-1| = 1, which is smaller than previous distance, so this is stored. This would be O(n).

**Can we make the problem smaller by splitting the input in half?**
Yes. The array won't change if we split in half, so this is okay.

**After splitting do I need BOTH halves or can I thrown one away?**
I need both halves because I need the whole array.

**What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?**
When there's only one element, that's the answer.

**What do we need to return for the recursion to work at the next level?**
Return the element with the smallest distance to 0.

**Is the input sorted or do we have indices to work with?**
Indices.

**How many recursive calls do we make? On what?**
We make 2 recursive calls on each half of the array.

**After the recursive calls return, what do I need to get the final answer?**
I need to return the element with the smallest distance to 0.

**What's my recurrence, and can I solve it with Master Theorem?**
$T(n) = 2T(n/2) + O(1)$, and Master Theorem can be used.

### 6.

**Problem:** Given a sorted array that has been rotated at some unknown pivot (e.g. [4,5,6,1,2,3]), find a target element x in O(log n) time.

**Example:** A = [4, 5, 6, 1, 2, 3], x = 1 Output: index 4 (1-based)

**Can we make the problem smaller by splitting the input in half?**
Yes, we only need the part of the array that contains the index.

**After splitting do I need BOTH halves or can I thrown one away?**\
I can throw one away. After splitting in half, we only need the half that can contain the target. So we check which half is sorted (one must be sorted) and see if our target is inside the range, and if so, recurse on it. If not, recurse on the other one.

**What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?**
If there's only one element, it either is the target or the target doesn't exist.

**What do we need to return for the recursion to work at the next level?**
Return the index of the target or -1.

**Is the input sorted or do we have indices to work with?**
Indices.

**How many recursive calls do we make? On what?**
We make one recursive call on half the array.

**After the recursive calls return, what do I need to get the final answer?**
Nothing. Whatever we recurse on will get the answer on its base case.

**What's my recurrence, and can I solve it with Master Theorem?**
$T(n) = T(n/2) + O(1)$, and Master Theorem can be used.

### 7.

**Problem:** Given an array of n numbers, find the k-th largest element in O(n) time.

**Example:** A = [3, 1, 5, 2, 4], k = 2 Output: 4

Selection algorithm.

### 8.

**Problem:** Given n points on a line, find the point that minimizes the sum of distances to all other points in O(n) time.

**Example:** A = [1, 3, 7, 10, 15] Output: 7

Median of medians.

1. Given a unimodal array (strictly increases then strictly decreases) and a target value x, determine if x exists in the array in O(log n) time.
2. Given a sorted array rotated at an unknown pivot, find the minimum element in O(log n) time.
3. Given two unsorted arrays A and B each of size n, find the k-th smallest element among all 2n elements in O(n) time.
5. Given k unsorted arrays each of size m (total n = k·m elements), find the median of all n elements in O(n) time.
6. Given a sorted array, find the element closest to a given target value T in O(log n) time.
9. Solve these recurrences:

T(n) = 3T(n/3) + n²
T(n) = 2T(n/2) + n log n
T(n) = 5T(n/5) + n
T(n) = T(2n/3) + 1

10. Given a sorted array of n integers where every element appears exactly twice except one element that appears once, find the unique element in O(log n) time.
11. 10. Given an array of n integers where one element appears more than n/2 times (called the majority element), find it in O(n) time.
   
Alternative 1 — Nuts and Bolts (also called Lock and Key): Given n nuts and n bolts, where each nut matches exactly one bolt, find all matching pairs in O(n log n) time. You can only compare a nut to a bolt, not nut-to-nut or bolt-to-bolt.
Alternative 2 — Dutch National Flag (Dijkstra): Given an array with only three distinct values (e.g., 0, 1, 2), sort it in O(n) time using a single pass.
Alternative 3 — Egg Drop Problem: Given k eggs and a building of n floors, find the highest safe floor in the minimum number of drops.
Alternative 4 — Karatsuba: Given two n-digit numbers, multiply them faster than O(n²) using D&C.
Alternative 5 — Closest Pair of Points: Given n points in a plane, find the two closest points in O(n log n) time.

Design an algorithm that does the following. Given an input array A of length n containing
distinct positive integers, output an array B so that B[i] contains the number of values
j 2 {0, ..., n  1} so that j>i and 3 · A[j] > A[i].
For example, if the input were A = [6, 2, 4, 1, 7], then the output would be B = [2, 3, 1, 1, 0].
Your algorithm should run in time O(n log n). If it helps, you may assume that n is a power
of 2.

An array of n elements contains all but one of the integers from 1 to n + 1.
1. Give the best algorithm you can for determining which number is missing
if the array is sorted, and analyze its asymptotic worst-case running time.
2. Give the best algorithm you can for determining which number is missing
if the array is not sorted, and analyze its asymptotic worst-case running
time.

In an attempt to take over Earth, evil aliens have contaminated certain water supplies with a virus that transforms humans into flesh-craving zombies. To track down the aliens, the Center for Disease Control needs to determine the epicenters of the outbreak—which water supplies have been contaminated. There are n potentially infected cities C = {c1, c2, . . . , cn}, but the FBI is certain that only k cities have contaminated water supplies. Unfortunately, the only known test to determine the contamination of a city’s water supply is to serve some of that water to a human and see whether they turn ravenous. Several brave volunteers have offered to undergo such an experiment, but they are only willing to try their luck once. Each volunteer is willing to drink a single glass of water that mixes together samples of water from any subset C 0 ⊆ C of the n cities, which reveals whether at least one city in C 0 had contaminated water. Your goal is to use the fewest possible experiments (volunteers) in order to determine, for each city ci , whether its water was contaminated, under the assumption that exactly k cities have contaminated water. You can design each experiment based on the results of all preceding experiments.

Save the world by designing an algorithm to determine which k of the n cities have contaminated water supplies using O(k lg n) experiments. Describe and analyze your algorithm.

9.1-1 — Find the second smallest element among n elements using at most n + lg(n) - 2 comparisons. Describe and analyze your algorithm.
9.1-2 — Find both the minimum and maximum of n elements using at most ⌈3n/2⌉ - 2 comparisons. Describe and analyze your algorithm.
9.3-3 — Modify quicksort to run in worst-case O(n log n) time using the selection algorithm. Describe and analyze your algorithm.
9.3-6 — Given an array of n elements, find all k-quantiles (the k-1 values that divide the array into k equal parts) in O(n log k) time.
9.3-7 — Given a sorted array of n distinct numbers and a number k, find the k numbers closest to the median in O(n) time.
Problem 9-2 — Given n elements each with a weight, find the weighted median in O(n) time.
