# Asymptotic Notation

## big-O, big-$\Omega$, big-$\Theta$

**big-O** represents an asymptotic **upper bound**, which is the <ins>worst</ins> it could go.  You would say, "the algorithm runs in n-time at worst, or it's O(n).

Consider comparing two different functions $f(n)$ and $g(n)$. $f(n) = O(g(n))$ (that is, $f(n)$ is big-O of $g(n)$) if and only if there exist a constant c > 0 and a constant $n_0$, such that $f(n) \leq c \times g(n)$ for all $n > n_0$.

So, for example if $f(n) = 3n + 5$ and $g(n) = n$, then $f(n) = O(g(n))$, since $3n + 5 \leq c \times n$ for any $c \geq 4$.

**big-$\Omega$**, on the other hand, represents an asymptoptic **lower bound**, which is the <ins>best</ins> it could go. You would say, "the algorithm runs in n-time at best, or it's $\Omega$(n).

Consider comparing two different functions $f(n)$ and $g(n)$. $f(n) = \Omega(g(n))$ (that is, $f(n)$ is big-$\Omega$ of $g(n)$) if and only if there exist a constant c > 0 and a constant $n_0$, such that $f(n) \geq c \times g(n)$ for all $n > n_0$.

So, for example if $f(n) = 3n + 5$ and $g(n) = n$, then $f(n) = \Omega(g(n))$, since $3n + 5 \geq c \times n$ for any $c \leq 3$.


And now you might be wondering: *wait, but then isn't both the upper bound and the lower bound of $3n + 5 = n$?* And yes, you are right. This a special example, because many times the upper bound is not necessarily the lower bound as well.

We refer to this "special case" as **big-$\Theta$**, and it's not a lower or upper bound, but it means that is exactly *that* which it represents. In our example, $3n + 5$ is both bigger than or equal to AND lesser than or equal to $n$, and that is only possible if $3n + 5$ is equal to $n$, which sounds weird to say, but since this is asymptotic, $3n + 5$ is actually super close to $n$, which to us means that it is so, so close, that it *basically* is $n$, or, in better words, "the algorithm is $\Theta(n)$."

Consider comparing two different functions $f(n)$ and $g(n)$. $f(n) = \Theta(g(n))$ (that is, $f(n)$ is big-$\Theta$ of $g(n)$) if and only if there exist constants $c_1 > 0, c_2 > 0$ and $n_0$, such that $c_1 \times g(n) \leq f(n) \leq c_2 \times g(n)$ for all $n > n_0$.

So, for example if $f(n) = 3n + 5$ and $g(n) = n$, then $f(n) = \Theta(g(n))$, since $c_1 \times n \leq 3n + 5 \leq c_2 \times n$ for any $c_1 \leq 4$ and $c_2 \geq 3$.

## small-o, small-$\omega$

There also exist versions of big-O and big-$\Omega$ that don't strictly require that our function might be equal to our upper or lower bound, but only that is less than or greater than. Those are called **small-o** and **small-$\omega$**, and the formula for them is similar to their big contraparts, but only, as mentioned, requiring the comparison to be absolute rather than inclusive. These are not required, but optional (still, strongly recommended to know, especially small-$\omega$.
