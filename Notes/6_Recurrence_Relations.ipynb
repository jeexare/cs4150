# Recurrence Relations

Consider $T(n) = 2T(\frac{n}{2})+n$. This is the time function of a recursive algorithm, and it means that, to solve a problem of size n, I split it into 2 subproblems, each of size $\frac{n}{2}$, solve each one recursively, then spend $n$ time combining the results. This is what happens in merge sort:

- **Divide**: split array in half -> 2 subproblems of size $\frac{n}{2}$
- **Conquer**: sort each half recusrively -> that's the $2T(\frac{n}{2})$
- **Combine**: merge the two sorted halves -> that takes $O(n)$ time

**What actually happens in MergeSort?**

Let's say it runs in an array of size n = 8. It splits into two problems of size 4. Each of those splits into two problems of size 2. Each of those splits into two problems of size 1.

At size 1, in returns the single element: there's nothing to sort. This is the **base case**.

```
n = 8
├── n = 4
│   ├── n = 2
│   │   ├── n = 1  ← base case
│   │   └── n = 1  ← base case
│   └── n = 2
│       ├── n = 1
│       └── n = 1
└── n = 4
    ├── n = 2
    │   ├── n = 1
    │   └── n = 1
    └── n = 2
        ├── n = 1
        └── n = 1
```

**How many levels does this tree have when n = 8?** It has 4 levels, 3 splits.

**And why does it stop at n = 1?** Becase one single element is already sorted.

**At each level of the tree, how much total work is being done?**

At the root (level 0), we have one problem of size 8, so combining costs $8$.

At level 1, we have two problems of size 4, each costs 4, fo total is $4 + 4 = 8$.

At level 2, we have four problems of size 2, each costs 2, so total is $2 + 2 + 2 + 2 = 8$

At level 3, we have eight problems of size 1, each costs 1, so total is $1 + 1 + 1 + 1 + 1 + 1 + 1 + 1 = 8$

Since each level costs $n = 8$ total, and there are 4 levels, the total is $4 \times 8 = 32$.

**How many levels are in general, for an array of size n?** $log n$

## Method 1: Expansion

>
> **1.** Expand a few times until you see the pattern
>
> **2.** Write the general form in terms of k
>
> **3.** Find k by setting the argument equal to 1, then substitute
>

There are several methods to solve these recurrence relations and find the actual upper bound. One of them is the **expansion** method. The idea is simple: we keep substituting the recurrence into itself, until we can find a pattern.

We start with:

$T(n) = 2T(\frac{n}{2}) + n$

And we know then that:

$T(\frac{n}{2}) = 2T(\frac{n}{4}) + \frac{n}{2}$

So we substitute that in:

$T(n) = 2[2T(\frac{n}{4}) + \frac{n}{2}] + n$

$T(n) = 4T(\frac{n}{4}) + 2\frac{n}{2} + n$

$T(n) = 4T(\frac{n}{4}) + n + n$

$T(n) = 4T(\frac{n}{4}) + 2n$

We also know that:

$T(\frac{n}{4}) = 2T(\frac{n}{8}) + \frac{n}{4}$

So:

$T(n) = 4[2T(\frac{n}{8}) + \frac{n}{4}] + 2n$

$T(n) = 8T(\frac{n}{8}) + 4\frac{n}{4} + 2n$

$T(n) = 8T(\frac{n}{8}) + n + 2n$

$T(n) = 8T(\frac{n}{8}) + 3n$

**Can we find the pattern?**

$2T(\frac{n}{2}) + n$

$4T(\frac{n}{4}) + 2n$

$8T(\frac{n}{8}) + 3n$

We can see that the coefficient of $n$ keeps increasing. We can can call this coefficient k, and see:

When $k = 1$, then the coefficient of T is $2$, and the argument is $\frac{n}{2}$ 

When $k = 2$, then the coefficient of T is $4$, and the argument is $\frac{n}{4}$ 

When $k = 3$, then the coefficient of T is $8$, and the argument is $\frac{n}{8}$

So, how can we solve for these coefficients in terms of $k$?

$2^1 = 2$, $2^2 = 4$, $2^3 = 8$, so we know that the coefficients are in terms of $2^k$ and the arguments in terms of $\frac{n}{2^k}$.

**If after $k$ steps we have $2^k \times T(\frac{n}{2^k}) + kn$, what value of $k$ makes $\frac{n}{2^k} = 1?$**

$n = 2^k$

$k = \log n$

Then we can plug this in here:

$2^{\log n} \times T(\frac{n}{2^{\log n}}) + n \log n$

$n \times T(\frac{n}{n}) + n \log n$

$n \times T(1) + n \log n$

$n \times 1 + n \log n$

$n + n \log n$ = $O(n \log n)$

## Method 2: Recursion Tree

>
> __1__. Draw the tree, writing the combining cost at each node
>    
> __2__. Add up the total cost at each level
>    
> **3**. Multiply by the number of levels
>
A **recursion tree** is like a visual version of expansion. Instead of algebra, we draw the tree and add up the costs level by levbel.

The trees starts with the root, which represents the original problem of size $n$. At each node, we write the cost of the combining step at that node, but not the recursive calls. Then we branch into subproblems:

```
              T(n)                    cost: n
             /    \
          T(n/2)  T(n/2)             cost: n/2 + n/2 = n
          /  \    /  \
       T(n/4)...T(n/4)               cost: n/4+n/4+n/4+n/4 = n
       ...
      T(1)T(1)...T(1)                cost: n (n leaves)
```

Similar to what we explained before, this tree shows that the total cost is $n log n$, because each level costs $n$, and we have $log n$ levels.

## Method 3: Guess and Verification

This method connects directly back to asymptotic notation proofs:

> **1.** Guess the answer.
>
> **2.** Verify it using the Big O definition
>

For example, for $T(n) = 2T(\frac{n}{2}) + n$, we guess that $T(n) = O(n log n)$, so we need to prove that $T(n) \leq c \times n log n$..

Let's assume $T(n/2) \leq c \times (\frac{n}{2}) \times \log (\frac{n}{2})$. Then we can substitute into the function, so:

$\leq 2 \times [c \times (\frac{n}{2}) \times \log (\frac{n}{2})] + n$

$= c \times n \times \log (\frac{n}{2}) + n$

To solve for $\log (\frac{n}{2})$ we remember:

$\log (\frac{n}{2}) = \log n - \log 2 = \log n - 1$

$= c \times n \times (\log n - 1) + n$

$= c \times n \log n - c \times n + n \leq c \times n \log n$

$=- c \times n + n \leq 0$

$= n \leq c \times n$

$= 1 \leq c $

$= c \geq 1 $

## Method 4: Masther Theorem

The **master theorem** aims to solve recurrences of the form $T(n) = aT(\frac{n}{b}) + f(n)$, where $a \geq 1, b > 1$ and $f(n)$ is the work done outside the recursive calls (usually, the combination step.

The **key quantity** to compute is $n^{\log_b a}$, that represents the "weight" of the recursive calls. Then, we compare $f(n)$ against it.

**Case 1:** When we compare $f(n)$ to our key quantity, the _recursion dominates_. This means that $f(n) = O(n^{\log_b a - \epsilon})$ for some $\epsilon > 0$, then $T(n) = \Theta(n^{\log_b a})$

**Case 2:** The function and our key quantity _are equal_, which means that if $T(n) = \Theta(n^{\log_b a})$, then $T(n) = \Theta(n^{\log_ba} \times log n)$.

**Case 3:** _Top-level work dominates_. If $f(n) = \Omega(n^{\log_ba + \epsilon})$ for some $\epsilon > 0$, AND $a \times f(\frac{n}{b}) \leq c \times f(n)$ for some $c < 1$, then $T(n) = \Theta(f(n))$.

Considering MergeSort, with the reccurence $T(n) = 2T(\frac{n}{2}) + n$.

$a = 2, b = 2, f(n) = n$

Our key quantity results in: $n^{\log_b a} = n^{\log_2 2} = n^1 = n$

And comparing our function to it gives us: $n = \Theta(n)$, which is case 2. 

So,

$T(n) = \Theta(n^{\log_b a} \log n) = \Theta(n \log n)$

Intuition: The work per level is the same ($n$) across all $log n$ levels of the recursion tree, so we multiply: $n \times log n$.
