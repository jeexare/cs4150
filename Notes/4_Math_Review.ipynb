# Math Review

Even though Math is not necessarily a favorite subject, it can be incredibly helpful and useful when understanding algorithmms. After all, when analyzing the time complexity of an algorithm, we use Math terminology. Some examples are:

$1$ --- Constant

$n$ --- Linear

$log n$ --- Logarithmic

$n^2$ --- Quadratic/Polynomial

$2^n$ --- Exponential

So, it's important to remember some basic rules:

## Exponentials

Exponents are just repeated multiplication. In this same fashion, negative exponents flip to division.

Consider, for all real $a > 0, m,$ and $n$,

$a^0 = 1$, because that's just how it is, no explanation

$a^1 = a$, because it's not multiplying $a$ with anything, it's just $a$

$a^{-1} = \frac{1}{a}$, because, as mentioned, it flips to division

$(a^m)^n = a^{mn}$, because raising a number to an exponent and then another is just the same as raising the number to the multiplication of the exponents... which is a useless explanation, because it has been exemplified already

## Logarithms

The key to remember this is to remember that logarithms are the _inverse_ of exponents. So, $\log_b a$ is asking "what power do I raise $b$ to get $a$?

Consider, for all real $a, b, c > 0$ and $n$,

$\log_c(ab) = \log_ca + \log_cb$. This is a basic/fundamental property, and it means that the logarithm of a product is the sum of the 2 logarithms.

$\log_ba^n = n\log_ba$. This is another property, stating that the logarithm of a number raised to a power is equal to "the exponent ($n$) times the logarithmn of the number ($\log_ba$).

$\log_ba = \frac{\log_ca}{\log_cb}$. This is the change of base formula. Since logarithms are basically about finding powers, if we want to find $\log_ba$, we can just use a different base $c$ and adjust it. This is useful is $c$ it's easier than $b$, for example.

$\log_b\frac{1}{a} = -\log_ba$. Very similar to its corresponding exponential rule. Same reasoning.

$a = b^{\log_b a}$. This is just substitution. If $y = \log_ba$, then $b^y = a$, so in a equation like $a = b^{\log_ba}$, $y$ is replaced by $\log_b a$.

## Arithmetic series

An arithmetic series is the sum of the terms in an arithmetic sequence. Each term after the first adds a constant. This is the formula of the first $n$ natural elements:

$$\sum_{k=1}^{n} k = 1 + 2 + \cdots + n = \frac{1}{2}n(n+1) = O(n^2).$$

## Sums of squares

This the formula for the sum of squares of the first $n$ natural numbers:

$$\sum_{k=1}^{n} k^2 = \frac{1}{6}n(n+1)(2n+1) = O(n^3).$$

This formula was derivied from a more complex derivation... but is compact enough to be memorized.

# Geometric series:

$$\sum_{k=0}^{n} x^k = 1 + x + x^2 + \cdots + x^n = \frac{1-x^{n+1}}{1-x}.$$

When the summation is infinite and $|x| < 1$, we have the infinite decreasing geometric series

$$\sum_{k=0}^{\infty} x^k = 1 + x + x^2 + \cdots = \frac{1}{1-x} = O(1).$$

For example:

$$\sum_{k=0}^{\infty} (1/2)^k = 1 + (1/2) + (1/2)^2 + \cdots = \frac{1}{1-(1/2)} = 2.$$

$$\sum_{k=0}^{\infty} (1/3)^k = 1 + (1/3) + (1/3)^2 + \cdots = \frac{1}{1-(1/3)} = 3/2.$$
