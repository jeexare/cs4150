# Algorithm Design Framework

When asked to design an algorithm, design first a **brute-force** algorithm, and ask the following questions:

1. Can we make the problem smaller by splitting the input in half?
2. After splitting do I need BOTH halves or can I thrown one away?
3. What's the smallest possible input where the answer id obvious or can be brute-forced with minimal cost?
4. What do we need to return for the recursion to work at the next level?
5. Is the input sorted or do we have indices to work with?
6. How many recursive calls do we make? On what?
7. After the recursive calls return, what do I need to get the final answer?
8. What's my recurrence, and can I solve it with Master Theorem?

## Practice problems
**Problem:** Given a sorted array A of n distinct integers, some of which may be negative, find an index i such that A[i] = i. If no such index exists, return -1.

**Example:** A = [-3, 0, 2, 4, 7, 9] The answer is i = 4 because A[4] = 4.

First, we can design a **brute-force** approach:

We loop through the array once, checking if A[i] is negative, then we go to the next one because i obviously not negative, and if we find a A[i] = i, we return i, otherwise we return -1.

This would O(n).

**Can we make the problem smaller by splitting the input in half?**
Yes, if we have an array |A| / 2, we can recurse on it and it can still work.

**After splitting do I need BOTH halves or can I thrown one away?**
I can throw one away, because the array is sorted, so if mid < A[mid], we know the answer is either on the left, or there's not answer, and viceversa.

**What's the smallest possible input where the answer is obvious or can be brute-forced with minimal cost?**
When there's only one element (i == j), we can check if A[i] == i.

**What do we need to return for the recursion to work at the next level?**
The index i, or -1.

**Is the input sorted or do we have indices to work with?**
Sorted.

**How many recursive calls do we make? On what?**
We can make 1 on only half the array each time, until the base case.

**After the recursive calls return, what do I need to get the final answer?**
We don't need to combine anything, just return the recursive call, since once it reaches the base case, we have our answer.

**What's my recurrence, and can I solve it with Master Theorem?**
$T(n) = T(n/2) + O(1)$, yes it can be solved with Master Theorem.

```
Algorithm FindUniqueIndex(A, i, j)
Input: A sorted array A with n distinct integers, a starting index i, an ending index j
Output: A index x such that A[x] == x; or -1

if i == j:
    if A[i] == i:
        return i
    else:
        return -1

mid = (i + j)/2

if A[mid] == mid:
    return mid
else if A[mid] < mid:
    return FindUniqueIndex(A, i, mid - 1)
else:
    return FindUniqueIndex(A, mid + 1, j)
```

**Problem:** Given a sorted array of n distinct positive integers, find if a number x exists in the array. Return true or false.

**Example:** A = [1, 3, 5, 7, 9, 11], x = 7 Output: true

**Problem:** Given a sorted array A of n integers, count how many elements are negative. Design an O(log n) algorithm.

**Example:** A = [-5, -3, -1, 2, 4, 7, 9] Output: 3

**Problem:** Given two sorted arrays A and B each of size n, find the median of all 2n elements combined in O(log n) time.

**Example:** A = [1, 3, 5], B = [2, 4, 6] Combined sorted: [1, 2, 3, 4, 5, 6] Output: 3 (or 4, middle elements)

**Problem:** Given an array of n integers, find the closest number to zero. Design an O(n) algorithm using D&C.

**Example:** A = [4, -1, 8, -6, 3, -2] Output: -1

**Problem:** Given a sorted array that has been rotated at some unknown pivot (e.g. [4,5,6,1,2,3]), find a target element x in O(log n) time.

**Example:** A = [4, 5, 6, 1, 2, 3], x = 1 Output: index 4 (1-based)

**Problem:** Given an array of n numbers where every element appears twice except one, find the unique element in O(n) time.

**Example:** A = [3, 1, 4, 1, 3, 2, 2] Output: 4

**Problem:** Given an array of n numbers, find the k-th largest element in O(n) time.

**Example:** A = [3, 1, 5, 2, 4], k = 2 Output: 4

**Problem:** Given n points on a line, find the point that minimizes the sum of distances to all other points in O(n) time.

**Example:** A = [1, 3, 7, 10, 15] Output: 7

**Problem:** Given a sorted array, find two elements that sum to a target value T in O(log n) time.

**Example:** A = [1, 3, 5, 7, 9, 11], T = 14
Output: (3, 11) or indices (2, 6)
