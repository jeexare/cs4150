# Data Structures

## Array

An array is a contiguous block of memory. It stores elements at indexed positions, which means that each element in an array of $n$ elements is stored at a position fronm $1$ to $n$.

Top thing to remember: For an index $i$ an an array $A$, we can access $A[1]$ in $O(1)$ time, because the memory address is computes as baseAddress $+ 1 *$ elementSize, so no traversal is neither. At the same time, inserting or deleting will be $O(n)$, because we have to shift elements

## LinkedLists

A linked list is a sequence of nodes. Each node holds a value and a pointer to the next node. The list starts at a **head** pointer:
```
head -> [val|->] -> [val|->] -> [val|->] -> null
```

The top thing to remember: insertion and deletion at a know position is $O(1)$ because the pointers just have to be "rewired" (make them point to the new next thing), but finding a value's position will take $O(n)$ because we have to traverse from the head.

## Queue (FIFO)

AKA 'first in, first out.' Similar to a line at a store, the first person in is the first person out. It takes $O(1)$ to both add (enqueue) and remove (dequeue), since it just adds at the end and removes from the front.

## Stack (FILO)

AKA 'first in, last out.' Similar to literally stacking object vertically, we always take first the objects at the top, which similarly were the last ones to be added. Add = push, remove = pop.

## Hash Tables

A hash table maps keys to values using a hash function that converts a key into an array index. Technically, it's just an array, so the important thing is the hash function andf the mapping.

Thed average case is $O(1)$ for searching, inserting, and deleting... but collision can make the perfomance degrade.

**Collision:** When two different keys hash to the same index. We need a good hash function to avoid this.

## Heaps and Priority Queues

Imagine a hospital emergency room (go watch The Pitt!): patients are always arriving, but we need to treat the most critical patient first, and that's not necesessarily whoever arrived first. So, we create a structure where we can always instantly grab the most urgent patient and easily add new ones. That's a heap.

A **min-heap**, especifically, always lets us grab the smallest element instantly. To build it, we follow two rules:

1. Every parent is smaller than its children
2. Every level is fully filled except possibly the last, which fills left to right. The height therefore is $O(log n)$

For example: 7, 9, 224, 26, 51, 35, 25
```
              7
            /   \
           9    24
          / \   / \
        26  51 35  25
```

Note, than a heap is not fully sorted. So, in this example, 25 is to the right of 35 even though 25 < 35, which is okay because we only care about order top to bottom, not left to right.

To store a heap, we don't even need a tree with pointers, and we can just use a plain array. The structure is implied by the indices:
```
Index:  1    2    3    4    5    6    7
      [ 7  | 9  | 24 | 26 | 51 | 35 | 25 ]
```

And the navigation formulas are very straightforward:
```
left child of i   →  2i
right child of i  →  2i + 1
parent of i       →  floor(i/2)
```



### Heap Operations

   ```
              7
            /   \
           9    24

        20  51 35  25
        /

    ```

In this operation we remove the, but obviously we can't just delete in leave a hole, so we have to shift things.
1. Swap root with the last element:
   ```
              25

    ```

#### BuildHeap

If we're givingh an unsorted array, how do we turn it into a heap?

```
Index:  1    2    3    4    5    6    7
      [ 95 | 10 | 15 | 25 | 20 | 30 | 40 ]


              95
            /    \
          10      15
         /  \    /  \
        25  20  30   40
```

The naive way wouldd be to insert each element one by one, but this is $O(n log n)$. A better way would be to start from the last non-leaf node (index = n/2), and then percolate down each node going right to left, bottom to top.

Why? Well, because nodes fdrom n/2+1 to n are all leaves, so they have no children, which means there's nothing to percolate down. Step by step:

n = 7, n/2 = 3

1 = 3 (15), children are 2i = 6 (30) and 2i + 1 = 7 (40). 15 < both, so no swap is needed

i = 2 (10), 2i = 4 (25) and 2i+1 = 5 (20), 10 < both, so no swap is needed

i = 1 (95), 2i = 2 (10) and 2i+1 = 3 (15), 95 > 10, swap with smaller child (10)

```
             10
            /    \
          95      15
         /  \    /  \
        25  20  30   40
```

Now 95 is index 2, children are 25 and 20. 95 > 20, swap with smaller child:

```
              10
            /    \
          20      15
         /  \    /  \
        25  95  30   40
```
Since 95 is now a leaf, we can stop.

The key insight is that most nodes are near the bottom and barely move.

### heapSort

HeapSort is just two steps:

1. Build a heap from n elements     →  O(n)
2. Call deleteMin n times           →  O(n log n)

We want to sort this array: [15, 10, 40, 25, 5, 30, 20]

Draw it as a tree:

```
              15
            /    \
          10      40
         /  \    /  \
        25   5  30   20
```

We start at i = floor(n/2) = 3, and **percolate down** going right-to-left

1 = 3 (40): children are 30 and 20 -> swap with 20

```
              15
            /    \
          10      20
         /  \    /  \
        25   5  30   40
```

1 = 2 (10): children are 25 and 5 -> swap with 5

```
              15
            /    \
           5      20
         /  \    /  \
        25  10  30   40
```

i = 1: (15) children are 5 and 20 -> swap with 5

Now, 15 drops to index 2: children are 25 and 10 -> swap with 10

```
              5
            /    \
          10      20
         /  \    /  \
        25  15  30   40
```

Each deleteMin should swap root with last element, shrink the heap by 1, and then percolate down. The extracted minimuns come out in sorted order.

startingHeap: [ 5  | 10 | 20 | 25 | 15 | 30 | 40 ]

**deleteMin 1**: extract 5. Swap root with last (40), percolate 40 down:

heap: [ 10 | 15 | 20 | 25 | 40 | 30 ], sorted: [5]

**deleteMin 2**: extract 10. Swap root with 30, percolate down.

heap: [ 1 | 15 | 20 | 25 | 40 ], sorted: [5, 10]

### BinarySearch Tree

Imagine a tree where every node has three things: a key, a left node, and a right node. The tule that makes it a search tree and not a regular binary tree is that:

**Everything in the left subtree < key < everything in nthe right subtree.**

So, for every node, all smaller values are to the left, and all the larger ones are to the right. For example:

```
              40
            /    \
          15      60
         /  \    /  \
        6   25  50   70
         \  / \
          9 20  30
         /    \
        8      22
```

In a BST, height matters a lot. If we define $n$ as the number of nodes and $h$ as the height, all of the BST operations hav to run in $O(h)$ time. For a balanced BST, $h = O(\log n)$, which is perfect. For a terrible unbalance tree, $h$ could even be $O(n)$, which is definitely not better than a linked list and not what we want.

#### Search

There are two versions of search, and both of them are $O(h)$.

Iterative version:

```
search(T, x):
    v = T.root
    while v != null:
        if v.key = x:
            return v
        if x < v.key:
            v = v.left
        if x > v.key:
            v = v.right
    return null
```

Recursive version:

```
search(v, x):
    if v = null:
        return null
    if v.key = x:
        return v
    if v.key < x:
        return search(v.left, x)
    if v.key > x:
        return search(v.right, x)
```

#### Insert

To insert, we search for where x is supposed to be, and then palce it there as a new leaf. For example, if we wanted to insert 17 into our example tree above, we'd "walk" 40 -> 15 -> 25 -> 20, and make 17 20's left child.

#### Delete

Three cases:

- Node is a leaf: we just remove it
- Node has one child: we splice it and connect the parent to the child
- Node has two children: we find the in-order successor (smallest node in right subtree), copy its key into the node, then delete the successor

For example, in case 1, using the example tree, let's say we wanted to delete 8. Since it has no children, we can literally just delete it with no issues. Now, let's say we wanted to delete 6: it has one child (9). We have to splice it out, by making 15 point to 9 instead of 6. Now, if we wanted to delete 15, this is where it gets tricky: we can't just remove it or splice it, as it has two subtrees that we don't want to lose or misplace. The fix is to find 15's in order successor, which is trhe smallest node in its right subtree. In this case, it's 20.

We copy 20's key into 15's spot, then delete 20 where it was. Since it only has one child, we just use case 2 and keep going.


### Augmenting BSTs

Imagine now we want to find the k-th smallest element efficiently. Like "give me the 3rd smallest number." A plain BST can't do this faster than O(n), so we would have to traverse everything, UNLESS we add a size attribute to each node, so it holds the number of nodes in the subtree rooted there. For example:

```
              40  [13]
            /        \
          15 [9]     60 [3]
         /    \      /    \
       6 [4]  25[4] 50[1] 70[1]
        \     / \
        9[2] 20[2] 30[1]
        /      \
       8[1]    22[1]
```

The formula that maintains it is:

```v.size = v.left.size + v.right.size + 1```


### Dynamic-Select

At any node then, the left subtree has ```m = v.left.size``` nodes, so the current node is the (m+1)-th smallest. That tells us where to go when searching:

- If ```k = m+1```, we return v
- If ```k < m+1```, we go left
- If ```k > m+1```, we go right, but now we search for `k-m-1`-th because we already "used up" m+1 elements

```
dynamicSelect(v, k):
    if v = null:
        return null
    m = v.left.size

    if k = m+1:
        return v
    if k < m+1:
        return dynamicSelect(v.left, k)
    if k > m+1:
        return dynamicSelect(v.right, k-m-1)
```

**Does adding v.size break insert/delete performance?**

Nope, it doesn't, because the Main Theorem guarantees that:

If an augmented attribute `f` for each node `v` depends only on `v`, `v.left`, and `v.right`, we can maintain `f` during insertions and delations without affecting the time complexity.


### BST Traversals

There are three ways to walk a tree, and all of them are $O(n)$. But the different lies in **when** we visit the current node relative to the children.

**Pre-order:** visit -> left -> right

**In-order:** left -> visit -> right

**Post-order:** left -> right -> visit

For example:

```
        4
       / \
      2   5
     / \
    1   3
```

**Pre-order**: 4, 2, 1, 3, 5

**In-order**: 1, 2, 3, 4, 5

**Post-order**: 1, 3, 2, 5, 4

Why do we use them?

Well, in-order is the most important one because it always gives a sorted list:

```
inOrder(v):
    if v = null:
        return
    inOrder(v.left)
    visit v
    inOrder(v.right)
```

Post-order is what we can use to compute the size of a node, because by the time we visit a node, both children already have their sizes:

```
postOrder(v):
    if v = null:
        return 0
    v.size = 1 + portOrder(v.left) + postOrder(v.right)
    return v.size
```

#### Range Minimum

Imagine we are given a BST where each node has a data value, and a range, and we're asked to find the node whose data is minimum among all the nodes whose keys fall in the range. The range itself doesn't have to keys in the tree.

We can add a min attribute that holds the minimum data value in the entire subtree rooted at v:

```v.min = min(v.data, v.left.min, v.right.min)```

This satisfies the main theorem because it depends only on v, v.left, and v.right.

1. We find the LCA
The LCA is the Lowest Common Ancestor, and in our range (xl, xr) is the highest node whose key falls inside the range. It's also then the node where the search paths for xl and xr split (to the left for xl and to the right for xr), an everything in the range lives below that node.

```
LCA(T, xl, xr):
    v = T.root
    while v != null and v.key ∉ [xl, xr]:
        if v.key < xl:
            v = v.right
        else:
            v = v.left
    return v
```

2. Range-Min from the LCA
Once we have the LCA `u`, we search left toward xl and right toward xr, collecting the minimums along thw way. The key thing is that when traversing left, if a node's kkey is equal or bigger than xl, we can safely include that node's data and its entire subtree's min, and symmetrically going right.

```
RangeMin(T, xl, xr):
    u = LCA(T, xl, xr)
    mindata = u.data

    v = u.left
    while v != null:
        if v.key > xl:
            mindata =  min(mindata, v.data, v.right.min)
            v = v.left
        if v.key < xl:
            v = v.right
        if v.key = xl:
            mindata = min(mindata, v.data, v.right.min)
            break
    
    search symmetricaly on u.right

    return mindata
```
