# Divide and Conquer

This is a powerful and simple technique for designing algorithms. Based on San Tzu's: if the problem feels too big to solve directtly, break it into smaller pieces, solve each piece and combine the answers.

## The recipe

1. **Divide:** Split the problem into smaller subproblems of the same type.
2. **Conquer:** Solve each subproblem. If they're too big, recurse. If they're tiny, solve them directly (this is the base case).
3. **Combine:** Merge the subproblem solutions into a solution for the original problem.

## Finding the maximum

Given an array $A$ of $n$ size, how ti find the largest element?

1. **Divide**: Split $A$ down the middle into a left halft and a right half.
2. **Conquer**: Find the max of the left half recursively, and find the max of the right half recursively.
3. **Combine**: Return whichever of the two maxes is bigger.

```
Algorithm FindMax(A, i, j)
Input: An array A of n elements, a starting index i, and a last index j
Output: The maximum element of A

if i = j:                   // Base case: If there's only one element,
    return A[i]             // return it. O(1)

mid = (j + 1) / 2           // Divide in the middle. O(1)

maxLeft = FindMax(A, i, mid)       // Conquer. Find the max of each half,
maxRight = FindMax(A, mid + 1, j)  // recursively. 2T(n/2)

if maxLeft >= maxRight:         // Combine step.
    return maxLeft              // Return the bigger
else:                           // out of the 2.
    return maxRight             // O(1)
```

We get the time function $T(n) = 2T(n/2) + O(1)$.

## MergeSort

Given an array $A$ of $n$ size, how do we sort it?

1. **Divide:** Split $A$ in the middle into 2 halves.
2. **Conquer**: Sort each half recursively.
3. **Combine**: Merge the two sorted halves into one sorted array.

```
Algorithm MergeSort(A, i, j)
Input: An array A of size n, a starting index i, and an ending index j
Output: A sorted array A

if i = j:          // Base case: A single element is
    return A       // already sorted. O(1)

mid = (j + i)/2    // Dividing in the middle. O(1)

MergeSort(A, i, mid)  // Conquer. Sort each half
MergeSort(A, mid + 1, j) // recursively. 2T(n/2)

Merge(A, i, j, mid) // Combine

Algorithm Merge(A, i, mid, j)
Input: An array A of size n, a starting index i, a middle index mid, and an ending index j
Output: A sorted array A

B = {} // And empty array to write O(1)
a = i // Beginning of left subarray O(1)
b = mid + 1 // Beginning of right subarray O(1)
c = 1 // Beginning of B, O(1)

while a <= mid AND b <=:                    // Run while there are still elements
    if A[a] <= A[b]:                        // on both subarrays. O(n)
        B[c] = A[a]
        a++
        c++
    else:
        B[c] = A[b]
        b++
        c++

if a <= mid:                           // Copy any leftover elements
    copy A[a...mid] to B[c..]          // O(n)
if b <= j
    copy A[b...j] to B[c...]

copy B to A[i...j]                    // Copying B back to A. O(n)
```

We get the time function $T(n) = 2T(n/2) + 2$.

## Maximum Subarray

Given an array $A$ with $n$ real numbers, how do we find the contiguous array with the largest sum?

1. **Divide**: Split in the middle
2. **Conquer**: Find the maximum subarray in each half recursively
3. **Combine**: Find the maximum subarray that crosses both halves, and return the maximum of the 3

```
Algorithm MaxSubarray(A, i, j)
Input: An array A of size n, a starting index i, and an ending index j
Output: The sum of the maximum contiguous subarray

if i = j:          // Base case: If there's only one element,
    return A       // that's the maximum subarray. O(i)

mid = (j+i)/2      // Dividing in the middle. O(1)

maxLeft = MaxSubArray(A, i, mid)       // Conquer. Find the max of each half
maxRight = MaxSubArray(A, mid + i, j)  // recursively. 2T(n/2)
maxCross = MaxCross(A, i, mid, j)       // Combine

return the largest out of maxLeft, maxRight, and maxCross

MaxCross(A, i, mid, j)
Input: An array A of size n, a starting index i, a middle index mid, and an ending index j
Output: The largest possible sum of a subArray that includes the middle

rightSum = A[mid + 1] // Holds the best sum to the right
sum = rightSum        // Current sum

for k = mid + 2 to j:       // Go from middle to end, O(n/2)
    sum = sum + A[k]
    if sum > rightSum:
        rightSum = sum

leftSum = A[mid + 1] // Holds the best sum to the left
sum = leftSum        // Current sum

for k = mid - 1 to i:       // Go from middle to beginning, O(n/2)
    sum = sum + A[k]
    if sum > leftSum:
        leftSum = sum

return leftSum + rightSum
```

We get the time function $T(n) = 2T(n/2) + n)$.
